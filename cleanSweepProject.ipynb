# Data Formatting

### Extra-characters on a string.

```python
# Time to give format to the data. 
df
```

```python
# Starting from left to right, forst column to attack is Last_Name.

# Some last names has characters that are not valid for a name or last name, like "/", "...". To
# delete them the  str.strip() function will work, giving as arguments the extra characters detected.
# lstrip will cut the left side.
# rstrip the right side.
# strip from both sides.

#df['Last_Name'] =  df['Last_Name'].str.lstrip("...")
#df['Last_Name'] =  df['Last_Name'].str.lstrip("/")
#df['Last_Name'] =  df['Last_Name'].str.rstrip("_")
df['Last_Name'] =  df['Last_Name'].str.strip("123._/")
```

```python
# Verifying the Last_Name column.
df
```

### Extra-characters on digits.

```python
# Second column: Phone_Number.

# This column have more variety of characters that needs to be deleted, str.strip() it is not enough to deal with it.
# Only numbers will remain in this column.
# Regular expression patterns are commonly used to match any character character combination.
# '[^\d]'match any character that is not a digit.
# str.replace will replace the regex with nothing ('')
# regex=True allows the recognition of the regex on the replace method

df['Phone_Number']=df['Phone_Number'].str.replace('[^\d]','',regex=True)
```

```python
# Check if all extra characters are gone.
df
```

### Concatenating strings.

```python
# Time to give some format. First all values will be transformed into string using the apply function.

# apply() method is used to apply a function to each element of the selected DataFrame column. In this case, the function 
# applied is an anonymous lambda function (a short, one-line function) that converts each element x in the 'Phone_Number'
# column to a string using the str() function.
df.Phone_Number = df.Phone_Number.apply(lambda x: str(x))
```

```python
# Now, this lambda function is splitting the 'Phone_Number' into three parts based on character positions (0-2, 3-5, and 6-9)
# then concatenates these parts with hyphens ('-') to create a formatted phone number 'xxx-xxx-xxxx'
df.Phone_Number = df.Phone_Number.apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# Confirm the function was applied correctly.
df
```

### Non-pattern-match cleanup.

```python
# nan-- and -- values can be deleted with the methods from the previous steps, but what if the column have even more?
# A function to automate this will be the best solution, in combination with regular expressions.
def replace_invalid_numbers(df):
    # Define a regular expression pattern to match 'xxx-xxx-xxxx' format.
    pattern = r'^\d{3}-\d{3}-\d{4}$'

    # Use the 'str.match()' method with the regular expression to filter valid phone numbers.
    valid_numbers = df['Phone_Number'].str.match(pattern)

    # Replace invalid phone numbers with empty strings ('')
    df.loc[~valid_numbers, 'Phone_Number'] = ''

    return df
```

```python
# Call the replacie_invalid_number function and return it to the dataframe
df = replace_invalid_numbers(df)
```

```python
# Check the results.
df
```

### String separation.

```python
# Perfect. Moving on: the address column
# The format for the addres is street, state, zip code, like row number 3. So, 3 new columns will be created
# The address column will be splited in 3, using str.split(), the comma as the delimiter of the split, n=2 to
# look for 2 commas on the string, and expand=true so store each part of the address into street, state and ZIP code
df[['Street', 'State', 'ZIP_Code']] = df['Address'].str.split(',',n=2, expand=True)
```

```python
# Checking the results.
df
```

### Specific cleanup.

```python
# Next columns: Paying Customer. Using replace() will give format to only Y if yes, N if No.
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
# Verify the results.
df
```

```python
# Same approach for Do_Not_Contact Column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
# Check again the result.
df
```

### Final missing value cleaning.

```python
# Clean againg the NaN values with an empty character ('')
df = df.fillna('')
```

```python
# Verify the results
df
```

```python
# Finally, clean the 'N/a'.
df = df.replace('N/a', '')
df
```

### Not useful entries cleanup.

```python
# As this is a customer calling list people with 'Do_Not_Contact' answered as Yes will be deleted.
# Using a for loop, check throuhg the column the condition described, if true it will drop that row.
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
# Same logic for the 'Phone_Number' column. You can't contact someone who didn't gave their phone
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df
```

```python
# The last step will be reset the index label from the dataframe.
df = df.reset_index(drop=True)
df
```
